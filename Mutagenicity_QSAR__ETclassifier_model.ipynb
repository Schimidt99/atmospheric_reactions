
# Extra Tree Classfier model for mutagenicity

```python
# Genotox_QSAR.csv
import io
import pandas as pd

from google.colab import files

uploaded = files.upload()

# Get the actual filename from the uploaded dictionary keys
filename = list(uploaded.keys())[0]  # Assuming only one file is uploaded

# Use the correct filename to read the CSV
data = pd.read_csv(io.BytesIO(uploaded[filename]), sep=";", decimal=",", header=0, encoding='latin1')
```

```python
import warnings

# Para ignorar todos os warnings
warnings.filterwarnings("ignore")
```

```python
# matriz
data
```

Removing columns with many zeros

```python
columns_to_drop = ['Ipc', 'SMR_VSA8', 'SlogP_VSA9', 'fr_prisulfonamd'] # Replace with your column names
data = data.drop(columns=columns_to_drop, axis=1)
```

```python
# Print the column names to verify the exact target column name
print(data.columns)
```

```python
unique_labels = data['OVERALL '].unique()
print(unique_labels)
```

```python
#the occurrences of each value in the 'OVERALL' column

print(data['OVERALL '].value_counts())
```

```python
saida = data['OVERALL ']   # output model
print(saida.shape)
```

```python
X = data.drop(['CAS_Number', 'Formula ', 'SMILES',	'OVERALL '], axis=1)   #eliminating columns from the original data, keeping only the descriptors
print(X.shape)
```

```python
# Autoscale input matrix
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler(with_mean=True, with_std=True)

# Drop rows with any missing values in X before scaling
X_cleaned = X.dropna()

# Also drop the corresponding rows from saida to keep indices aligned
saida_cleaned = saida.loc[X_cleaned.index]

# Fit and transform the cleaned data
X2 = pd.DataFrame(scaler.fit_transform(X_cleaned), index=X_cleaned.index, columns=X_cleaned.columns) # Preserve original index and column names
```

```python
X2
```

```python
import sklearn
print(sklearn.__version__)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X2, saida_cleaned, test_size=0.20, random_state=13)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Modelo Extra Tree Classfier

```python
from sklearn.ensemble import ExtraTreesClassifier

modelo = ExtraTreesClassifier(n_estimators=100, min_samples_leaf=2, random_state=13).fit(X_train, y_train)
```

```python
# predict calculation

ypred = modelo.predict(X_test)
```

# Avaliation metrics

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
#multiclass avaliation

from sklearn.metrics import classification_report

report = classification_report(y_test, ypred)
print(report)


precision = precision_score(y_test, ypred, average='weighted')
print("The precision for the test set: {:.4f}".format(precision))

recall = recall_score(y_test, ypred, average='weighted')
print("The recall for the test set: {:.4f}".format(recall))

f1 = f1_score(y_test, ypred, average='weighted')
print("The f1 score for the test set: {:.4f}".format(f1))

accuracy = accuracy_score(y_test, ypred)
print("The accuracy for the test set: {:.4f}".format(accuracy))
```

```python
# Matriz de confusÃ£o

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, ypred)
print(cm)
```

```python
# How many samples for each 'label' in y_test
print(y_test.value_counts())
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[1, 2, 3])  # Replace with your class labels
disp.plot(cmap='Blues')
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import roc_auc_score

# Assuming y_test and y_pred_proba are appropriate for your multiclass problem
# For multiclass, you'll need predict_proba to output probabilities for each class
y_pred_proba_multiclass = modelo.predict_proba(X_test)

# Calculate multiclass ROC AUC using 'ovr' and 'weighted' averaging
roc_auc_multiclass = roc_auc_score(y_test, y_pred_proba_multiclass, multi_class='ovr', average='weighted')
print("Multiclass ROC AUC Score (OvR, Macro):", "{:.4f}".format(roc_auc_multiclass))
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Choose the number of top features you want to select
k = 70 # You can change this number based on your needs

# Create a SelectKBest object with the desired scoring function
selector = SelectKBest(score_func=f_classif, k=k)

# Fit and transform your data to select the top k features
X_new = selector.fit_transform(X2, saida_cleaned)

# Get the indices of the selected features
selected_features_indices = selector.get_support(indices=True)

# Get the names of the selected features
selected_features_names = X2.columns[selected_features_indices]

print("Original number of features:", X2.shape[1])
print("Number of selected features:", X_new.shape[1])
print("Names of selected features:", selected_features_names)
```
